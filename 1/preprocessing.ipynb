```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import  LabelEncoder
```

```python
df = pd.read_csv('ObesityDataSet.csv')
df
```

```python
#Preprocessing Awal
display(df.duplicated().sum())
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
# Outlier Remove
def remove_outliers_zscore(dataframe, column, threshold=3):
    mean = dataframe[column].mean()
    std = dataframe[column].std()
    z_scores = (dataframe[column] - mean) / std
    return dataframe[np.abs(z_scores) < threshold]
df = remove_outliers_zscore(df, 'Age')
df = remove_outliers_zscore(df, 'FCVC')
df = remove_outliers_zscore(df, 'NCP')
df = remove_outliers_zscore(df, 'CH2O') 
df = remove_outliers_zscore(df, 'FAF')
df = remove_outliers_zscore(df, 'TUE')
df.shape
```

```python
# Label Encodeer
df_label = df.copy()
le = LabelEncoder()
df_label["Gender"] = le.fit_transform(df_label["Gender"])
df_label["SMOKE"]=  le.fit_transform(df_label["SMOKE"])
df_label["FAVC"]=  le.fit_transform(df_label["FAVC"])
df_label["SCC"] = le.fit_transform(df_label["SCC"])
df_label["CALC"] = le.fit_transform(df_label["CALC"])
df_label["MTRANS"] = le.fit_transform(df_label["MTRANS"])
df_label["CAEC"] = le.fit_transform(df_label["CAEC"])
df_label['family_history_with_overweight'] = le.fit_transform(df_label['family_history_with_overweight'])
df_label["NObeyesdad"] = le.fit_transform(df_label["NObeyesdad"])

df_label
```

```python
df_label.to_csv('DataCleanLabel.csv', index=False)
```

```python
#One Hot Encoder Only
df_onehot = df.copy() 
col = ['Gender','SMOKE', 'FAVC', 'SCC', 'CAEC', 'CALC', 'MTRANS', 'family_history_with_overweight', 'NObeyesdad']
df_onehot = pd.get_dummies(df_onehot, columns=col, drop_first=True, dtype=int)
df_onehot
```

```python
df_onehot.to_csv("DataCleanOneHot.csv", index=False)
```
