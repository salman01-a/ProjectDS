```python
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder
```

```python
df = pd.read_csv('ObesityDataSet.csv')
df.head()
```

```python
df['BMI'] = df['Weight'] / (df['Height'] ** 2)
df['BMI'] = df['BMI'].round(2)
```

```python
#Encoding categorical variables
df = df.drop(columns="Gender")
df
```

```python
# Label Encodeer
le = LabelEncoder()
df["SMOKE"]=  le.fit_transform(df["SMOKE"])
df["FAVC"]=  le.fit_transform(df["FAVC"])
df["SCC"] = le.fit_transform(df["SCC"])
df["CALC"] = le.fit_transform(df["CALC"])
df["MTRANS"] = le.fit_transform(df["MTRANS"])
df["CAEC"] = le.fit_transform(df["CAEC"])
df['family_history_with_overweight'] = le.fit_transform(df['family_history_with_overweight'])

df
```

```python
#Outlier Detection and Removal
def remove_outliers_zscore(dataframe, threshold=3):
    numeric_cols = dataframe.select_dtypes(include=[np.number]).columns
    z_scores = np.abs((dataframe[numeric_cols] - dataframe[numeric_cols].mean()) / dataframe[numeric_cols].std())
    filtered_entries = (z_scores < threshold).all(axis=1)
    return dataframe[filtered_entries]
df_cleaned = remove_outliers_zscore(df)
df_cleaned.reset_index(drop=True, inplace=True)
df_cleaned

df_cleaned.to_csv('ObesityDataSet_Cleaned.csv', index=False)

```

```python
df= pd.read_csv('ObesityDataSet.csv')
df['BMI'] = df['Weight'] / (df['Height'] ** 2)
df['BMI'] = df['BMI'].round(2)
df = df.drop(columns="Gender")
```

```python
#Label Encoder
le = LabelEncoder()
df["SMOKE"]=  le.fit_transform(df["SMOKE"])
df["FAVC"]=  le.fit_transform(df["FAVC"])
df["SCC"] = le.fit_transform(df["SCC"])
df['family_history_with_overweight'] = le.fit_transform(df['family_history_with_overweight'])

#One Hot Encoder
one_hot_cols = ['CAEC', 'CALC', 'MTRANS']

# Lakukan One-Hot Encoding dan langsung gabungkan
df = pd.get_dummies(df, columns=one_hot_cols, drop_first=True, dtype=int)
df
```
